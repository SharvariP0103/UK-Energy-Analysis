```python
import matplotlib.pyplot as plt

df['MONTH_YEAR'] = df['DATE'].dt.to_period('M')
monthly_demand = df.groupby('MONTH_YEAR')['NET_DEMAND'].mean().reset_index()
monthly_demand['MONTH_YEAR'] = monthly_demand['MONTH_YEAR'].astype(str)

plt.figure(figsize=(8, 4))
plt.plot(monthly_demand['MONTH_YEAR'], monthly_demand['NET_DEMAND'])
plt.xlabel('Month-Year')
plt.ylabel('Average Net Demand')
plt.title('Average Net Demand Over Time (Monthly)')

# Filter for the 1st and 6th month of each year
filtered_months = monthly_demand[monthly_demand['MONTH_YEAR'].str.endswith(('-01', '-06'))]
plt.xticks(filtered_months.index, filtered_months['MONTH_YEAR'], rotation=45)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
monthly_generation = df.groupby('MONTH_YEAR')['GENERATION'].mean().reset_index()
monthly_generation['MONTH_YEAR'] = monthly_generation['MONTH_YEAR'].astype(str)

plt.figure(figsize=(8, 4))
plt.plot(monthly_generation['MONTH_YEAR'], monthly_generation['GENERATION'])
plt.xlabel('Month-Year')
plt.ylabel('Average Generation')
plt.title('Average Generation Over Time (Monthly)')

# Filter for the 1st and 6th month of each year
filtered_months = monthly_generation[monthly_generation['MONTH_YEAR'].str.endswith(('-01', '-06'))]
plt.xticks(filtered_months.index, filtered_months['MONTH_YEAR'], rotation=45)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
merged_data = pd.merge(monthly_demand, monthly_generation, on='MONTH_YEAR')

plt.figure(figsize=(8, 4))
plt.plot(merged_data['MONTH_YEAR'], merged_data['NET_DEMAND'], label='Average Net Demand')
plt.plot(merged_data['MONTH_YEAR'], merged_data['GENERATION'], label='Average Generation')

plt.xlabel('Month-Year')
plt.ylabel('Average Value')
plt.title('Comparison of Average Monthly Net Demand and Generation')

# Filter for the 1st and 6th month of each year
filtered_months = merged_data[merged_data['MONTH_YEAR'].str.endswith(('-01', '-06'))]
plt.xticks(filtered_months.index, filtered_months['MONTH_YEAR'], rotation=45)

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Task
Analyze the distribution of fuels in net generation and demand from the dataset "demand_and_generation_2017.csv".

## Calculate fuel contributions

### Subtask:
Calculate the percentage contribution of each fuel type to the total generation and net demand.


**Reasoning**:
Calculate the total generation, percentage contribution of each fuel type to total generation, and the contribution of each fuel type percentage to net demand, then display the head of the DataFrame.



```python
generation_columns = ['GAS', 'COAL', 'NUCLEAR', 'SCOT_WIND', 'ENG_WIND', 'WALES_NI_WIND', 'HYDRO', 'IMPORTS', 'BIOMASS', 'OTHER', 'SOLAR']

df['TOTAL_GENERATION'] = df[generation_columns].sum(axis=1)

for col in generation_columns:
    df[col + '_perc_calc'] = (df[col] / df['TOTAL_GENERATION']) * 100
    df[col + '_demand'] = (df[col + '_perc_calc'] / 100) * df['NET_DEMAND']

display(df.head())
```

## Visualize distribution

### Subtask:
Create visualizations (e.g., stacked bar charts or pie charts) to show the distribution of fuel types in both net generation and demand over time or as an overall average.


**Reasoning**:
Select the columns representing the percentage contribution of each fuel type to the total generation, calculate the overall average percentage contribution for each fuel type, and create a pie chart to visualize the average fuel distribution in total generation. Add a title to the pie chart.



```python
# Select columns related to fuel generation percentages
generation_perc_columns = [col for col in df.columns if col.endswith('_perc_calc') and col.split('_perc_calc')[0] in ['GAS', 'COAL', 'NUCLEAR', 'SCOT_WIND', 'ENG_WIND', 'WALES_NI_WIND', 'HYDRO', 'IMPORTS', 'BIOMASS', 'OTHER', 'SOLAR']]

# Calculate the overall average percentage contribution for each fuel type for generation
average_generation_perc = df[generation_perc_columns].mean()

# Create a pie chart for average fuel distribution in total generation
plt.figure(figsize=(8, 8)) # Increased figure size
plt.pie(average_generation_perc, labels=[col.replace('_perc_calc', '') for col in average_generation_perc.index], autopct='%1.1f%%', startangle=140, pctdistance=0.85) # Added labels, autopct, and pctdistance
plt.title('Average Fuel Distribution in Total Generation', pad=20) # Added padding
plt.axis('equal')

plt.show()
```

**Reasoning**:
Select the columns representing the estimated contribution of each fuel type's generation percentage to the net demand, calculate the overall average estimated demand met by each fuel type, and create a pie chart to visualize the average estimated fuel distribution in net demand. Add a title to the pie chart and display both pie charts.



```python
# Select columns related to estimated demand met by each fuel type
demand_met_columns = [col for col in df.columns if col.endswith('_demand') and col.split('_demand')[0] in ['GAS', 'COAL', 'NUCLEAR', 'SCOT_WIND', 'ENG_WIND', 'WALES_NI_WIND', 'HYDRO', 'IMPORTS', 'BIOMASS', 'OTHER', 'SOLAR']]

# Calculate the overall average estimated demand met by each fuel type
average_demand_met = df[demand_met_columns].mean()

# Create a pie chart for average estimated fuel distribution in net demand
plt.figure(figsize=(8, 8))
plt.pie(average_demand_met, labels=[col.replace('_demand', '') for col in average_demand_met.index], autopct='%1.1f%%', startangle=140, pctdistance=0.85) # Added labels, autopct, and pctdistance
plt.title('Average Estimated Fuel Distribution in Net Demand', pad=20) # Added padding
plt.axis('equal')

plt.show()
```

## Analyze trends

### Subtask:
Analyze the trends in fuel distribution over time to identify any significant shifts or patterns.


**Reasoning**:
Calculate the monthly average percentage of each fuel type in the total generation and the monthly average estimated contribution of each fuel type to the net demand, then prepare the data for plotting.



```python
# Calculate the monthly average percentage of each fuel type in the total generation
monthly_generation_perc = df.groupby('MONTH_YEAR')[generation_perc_columns].mean().reset_index()
monthly_generation_perc = monthly_generation_perc.set_index('MONTH_YEAR')

# Calculate the monthly average estimated contribution of each fuel type to the net demand
monthly_demand_met = df.groupby('MONTH_YEAR')[demand_met_columns].mean().reset_index()
monthly_demand_met = monthly_demand_met.set_index('MONTH_YEAR')
```

**Reasoning**:
Plot stacked bar charts showing the monthly average percentage distribution of fuel types in total generation over time and the monthly average estimated distribution of fuel types in net demand over time.



```python
# Calculate the overall average percentage contribution for each fuel type for generation to determine order
average_generation_perc_ordered = monthly_generation_perc.mean().sort_values(ascending=False)
ordered_generation_columns = average_generation_perc_ordered.index.tolist()

# Plot stacked area chart for monthly average percentage distribution of fuel types in total generation
plt.figure(figsize=(25, 25)) # Increased figure size significantly
monthly_generation_perc[ordered_generation_columns].plot(kind='area', stacked=True, cmap='tab20b')
plt.title('Monthly Average Percentage Distribution of Fuel Types in Total Generation')
plt.xlabel('Month-Year')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45, ha='right') # Show all months
plt.legend(title='Fuel Type', bbox_to_anchor=(1.05, 1), loc='upper left', labels=[col.replace('_perc_calc', '') for col in ordered_generation_columns]) # Ensure labels match plotted columns
plt.tight_layout()
plt.show()

# Calculate the overall average estimated contribution for each fuel type for demand to determine order
average_demand_met_ordered = monthly_demand_met.mean().sort_values(ascending=False)
ordered_demand_columns = average_demand_met_ordered.index.tolist()

# Plot stacked area chart for monthly average estimated distribution of fuel types in net demand
plt.figure(figsize=(15, 15))
monthly_demand_met[ordered_demand_columns].plot(kind='area', stacked=True, cmap='tab20b')
plt.title('Monthly Average Estimated Distribution of Fuel Types in Net Demand')
plt.xlabel('Month-Year')
plt.ylabel('Estimated Contribution to Net Demand')
plt.xticks(rotation=45, ha='right') # Show all months
plt.legend(title='Fuel Type', bbox_to_anchor=(1.05, 1), loc='upper left', labels=[col.replace('_demand', '') for col in ordered_demand_columns]) # Ensure labels match plotted columns
plt.tight_layout()
plt.show()
```

# Task
Forecast the demand and generation trend for the next 2 years using an Exponential Smoothing model.

## Model selection (exponential smoothing)

### Subtask:
Select a suitable Exponential Smoothing model (e.g., Holt-Winters) considering the seasonality and trend in the data.


**Reasoning**:
Analyze the time series plots for demand and generation to identify the presence of trend and seasonality and determine the frequency of the time series data to select a suitable Exponential Smoothing model.



```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Determine the frequency of the time series data
# Since the data is grouped by MONTH_YEAR, the frequency is monthly.
frequency = 'M'

# Decompose the monthly average demand time series
monthly_demand.index = pd.to_datetime(monthly_demand['MONTH_YEAR'])
decomposition_demand = seasonal_decompose(monthly_demand['NET_DEMAND'], model='additive', period=12)

# Plot the decomposed demand time series
fig_demand = decomposition_demand.plot()
fig_demand.set_size_inches(10, 8)
plt.suptitle('Seasonal Decomposition of Monthly Average Net Demand', y=1.02)
plt.show()

# Decompose the monthly average generation time series
monthly_generation.index = pd.to_datetime(monthly_generation['MONTH_YEAR'])
decomposition_generation = seasonal_decompose(monthly_generation['GENERATION'], model='additive', period=12)

# Plot the decomposed generation time series
fig_generation = decomposition_generation.plot()
fig_generation.set_size_inches(10, 8)
plt.suptitle('Seasonal Decomposition of Monthly Average Generation', y=1.02)
plt.show()

# Based on the decomposition plots, both demand and generation show a trend and seasonality.
# Therefore, Holt-Winters Exponential Smoothing is a suitable model.
# We will use the ExponentialSmoothing class from statsmodels with trend and seasonal components.
```

## Model training

### Subtask:
Train separate Exponential Smoothing models on the historical demand and generation data.


**Reasoning**:
Import the necessary class and train the Exponential Smoothing models for both demand and generation with specified parameters.



```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Instantiate and fit the Exponential Smoothing model for monthly average net demand
fitted_demand_model = ExponentialSmoothing(
    monthly_demand['NET_DEMAND'],
    seasonal_periods=12,
    trend='add',
    seasonal='add'
).fit()

# Instantiate and fit the Exponential Smoothing model for monthly average generation
fitted_generation_model = ExponentialSmoothing(
    monthly_generation['GENERATION'],
    seasonal_periods=12,
    trend='add',
    seasonal='add'
).fit()
```

## Forecasting

### Subtask:
Use the trained Exponential Smoothing models to forecast demand and generation for the next two years.


**Reasoning**:
Use the fitted Exponential Smoothing models to forecast demand and generation for the next two years.



```python
# Use the fitted demand model to generate forecasts for the next 24 months
forecast_demand = fitted_demand_model.forecast(40)

# Use the fitted generation model to generate forecasts for the next 24 months
forecast_generation = fitted_generation_model.forecast(40)
```

## Visualization

### Subtask:
Visualize the historical data and the forecasted trends.


**Reasoning**:
Create two plots, one for historical and forecasted demand and one for historical and forecasted generation, as instructed.



```python
# Create a figure and axes for the demand plot
plt.figure(figsize=(10, 6))
ax_demand = plt.gca()

# Plot historical monthly average net demand
monthly_demand['NET_DEMAND'].plot(ax=ax_demand, label='Historical Net Demand')

# Plot forecasted demand
forecast_demand.plot(ax=ax_demand, label='Forecasted Net Demand')

# Add title and labels to the demand plot
plt.title('Monthly Average Net Demand and 2-Year Forecast')
plt.xlabel('Month-Year')
plt.ylabel('Average Net Demand')

# Add legend to the demand plot
ax_demand.legend()

# Create a figure and axes for the generation plot
plt.figure(figsize=(10, 6))
ax_generation = plt.gca()

# Plot historical monthly average generation
monthly_generation['GENERATION'].plot(ax=ax_generation, label='Historical Generation')

# Plot forecasted generation
forecast_generation.plot(ax=ax_generation, label='Forecasted Generation')

# Add title and labels to the generation plot
plt.title('Monthly Average Generation and 2-Year Forecast')
plt.xlabel('Month-Year')
plt.ylabel('Average Generation')

# Add legend to the generation plot
ax_generation.legend()

# Display both plots
plt.tight_layout()
plt.show()
```
